```python
def showgraph(DDimpactNR):
    ImpactNR = DDimpactNR

    Impact,ImpactUnit = ImpactInfo(ImpactNR)

    ImpactEV = Impact + " EV"
    ImpactICEV = Impact + " ICEV"
    ImpactUnit = ImpactUnit + " / km"

    plots = pd.DataFrame()
    plots = pd.concat([EV[ImpactEV], ICEV[ImpactICEV]], axis=1, sort=False)
    plots.fillna(0,inplace=True)


    # To show grams for Climate Change instead of kg
    if ImpactNR == 2:
        plots = plots*1000
    # Only show if value is over 1% 
    for i in range(2):
        a = np.array(plots.iloc[:,i].values.tolist())
        plots.iloc[:,i] = np.where(a < (plots.iloc[:,1].sum()/100) , False, a).tolist()
    plots = plots[(plots.T != 0).any()]

    
    fig, ax = plt.subplots(figsize=(20,10))
    # ploting the 2 stacked bars
    EVplot = plots.T.plot.bar(
        stacked=True,
        ax = ax,
        colormap="Set1",
        width=0.5,  
    )

    height = plots.sum()
    plt.ylim(0,(max(height)*1.1))

    #Legend and grid
    ax.legend(loc=2,prop={'size':13})
    ax.grid(axis='y')
    ax.set_axisbelow(True)
    #Labels
    titlestring = Impact + " impact " + "(" + ImpactUnit + ")" + ". Showing values >1%"
    plt.title(titlestring,size=30)
    ax.set_ylabel(ImpactUnit,fontsize=20)
    plt.xticks(size=20,rotation=0)
    ax.set_xticklabels(['EV','ICEV'])

    # Show values on the bars

    horizontalALI = 'center'
    verticalALI = 'center'
    left = right = False
    offset = 0
    for rect, value in zip(ax.patches, plots.values.flatten()):
        if not value==0:

            h = rect.get_height() /2.
            w = rect.get_width() /2.
            x, y = rect.get_xy()

            height = plots.sum()
            prosenta = 3
            if x == -0.25:
                if h <= (height[:][0]/100)*prosenta:
                    if not left:
                        horizontalALI = 'left'
                        offset = 0.25
                        left = True
                    else:
                        horizontalALI = 'center'
                        offset = 0
                        left = False               
                else:
                    horizontalALI = 'center'
                    offset = 0
                    left = False

            elif x == 0.75:
                if h <= (height[:][1]/100)*prosenta:
                    if not right:
                        horizontalALI = 'left'
                        offset = 0.25
                        right = True
                    else:
                        horizontalALI = 'center'
                        offset = 0
                        right = False
                else:
                    horizontalALI = 'center'
                    offset = 0
                    right = False

            if value < 0.01:
                ax.text(x+w+offset, y+h,str('{:.2e}'.format(value)),horizontalalignment=horizontalALI,verticalalignment=verticalALI,fontsize = 15,rotation=0)
            else:
                ax.text(x+w+offset, y+h,str(round(value,3)),horizontalalignment=horizontalALI,verticalalignment=verticalALI,fontsize = 15,rotation=0)
    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

    # Put a legend to the right of the current axis
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5)) 
    #EVplot.set_visible(False)
    plt.ion()
    return fig

    #plt.show()
```

```python
def ImpactInfo(i):

    if i == 1:
        return('Acidification', 'molc H+ eq')
    elif i ==2:
        return('Climate change','g CO2 eq')
    elif i == 3:
        return('Freshwater ecotoxicity','CTUe')
    elif i == 4:
        return('Freshwater eutrophication','kg P eq')
    elif i == 5:
        return('Human toxicity, cancer effects','CTUh')
    elif i == 6:
        return('Human toxicity, non-cancer effects','CTUh')
    elif i == 7:
        return('Ionizing radiation E (interim)','CTUe')
    elif i == 8:
        return('Ionizing radiation HH','kBq U235 eq')
    elif i == 9:
        return ('Land use','kg C deficit')
    elif i == 10:
        return('Marine eutrophication','kg N eq')
    elif i == 11:
        return('Mineral, fossil & ren resource depletion','kg Sb eq')
    elif i == 12:
        return('Ozone depletion','kg CFC-11 eq')
    elif i == 13:
        return('Particulate matter','kg PM2.5 eq')
    elif i == 14:
        return('Photochemical ozone formation','kg NMVOC eq')
    elif i == 15:
        return('Terrestrial eutrophication','molc N eq')
    elif i == 16:
        return('Water resource depletion','m3 water eq')
```
